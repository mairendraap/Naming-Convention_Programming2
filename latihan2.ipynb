```python
class dtBrg:
    """Merepresentasikan data sebuah barang atau produk."""

    def _init_(self, n, hrg):
        """Membuat objek barang dengan nama dan harga tertentu.

        Parameters:
        n : nama barang
        hrg: harga barang
        """
        self.NM = n
        self.HRG = hrg
        
    def hitung Disc(self, tp):
        """Menghitung diskon berdasarkan tipe pelanggan.

        Parameters:
        tp tipe pelanggan
        """
        if tp == "gold":
            return self. HRG * 0.2
        elif tp == "silver":
            return self. HRG * 0.1
        else:
            return 0

def CETAK(d, tP):
    """Menampilkan rincian pembayaran barang.
    
    Parameters:
    d : objek barang
    tP: tipe pelanggan
    """
    ds = d.hitung Disc(tP)
    ttl = d.HRG - ds

    print("Nama Barang :", d.NM)
    print("Harga :", d.HRG)
    print("Diskon :", ds)
    print("Total:", ttl)

x = dtBrg("Laptop", 10000000)
CETAK(x, "gold")
```

```python
class DataBarang:
    """Merepresentasikan data sebuah barang atau produk."""

    def __init__(self, nama, harga):
        """Membuat objek barang dengan nama dan harga tertentu.

        Parameters:
        nama : nama barang
        harga: harga barang
        """
        self.nama = nama
        self.harga = harga

    def hitung_diskon(self, tipe_pelanggan):
        """Menghitung diskon berdasarkan tipe pelanggan.

        Parameters:
        tipe_pelanggan : tipe pelanggan
        """
        if tipe_pelanggan == "gold":
            return self.harga * 0.2
        elif tipe_pelanggan == "silver":
            return self.harga * 0.1
        else:
            return 0


def cetak_rincian(barang, tipe_pelanggan):
    """Menampilkan rincian pembayaran barang.

    Parameters:
    barang : objek barang
    tipe_pelanggan: tipe pelanggan
    """
    diskon = barang.hitung_diskon(tipe_pelanggan)
    total = barang.harga - diskon

    print("Nama Barang :", barang.nama)
    print("Harga       :", barang.harga)
    print("Diskon      :", diskon)
    print("Total       :", total)


# Eksekusi Utama
laptop = DataBarang("Laptop", 10000000)
cetak_rincian(laptop, "gold")
```
